# Import all modules

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

# Import data and display it

```python
df = pd.read_pickle('power_consumption')
df.describe()
```

```python
df.head()
```

# Index the data with the column Datetime

```python
df.index = df['Datetime']
df.head()
```

# Import the Calendar module and add day names to the dataset

```python
import calendar
df['Day_name'] = [calendar.day_name[x] for x in df.Datetime.dt.dayofweek]
df.head()
```

# Extract one week worth of data from that main data set and display

```python
week = df['2007-01-01':'2007-01-07'] # 1 week of readings
week.head()
```

# Plot the 2 hour average for the data from the week

```python
week.resample('2H').mean()['Global_intensity'].plot() # resampled to one value every 2 hours
```

# Plot that data from the week with the day names as the x-axis

```python
# A more complex plot, with weekdays shown on the graph
week.plot(x='Day_name', y='Global_intensity', figsize=(12, 6), 
                xticks = range(0,  len(week.index), len(week.index)/7)) # Little bit hacky, but gets us all the days
```

# Locate all the data within the weeks data that has Thursday in Dayname
###What is interesting to note is that the locating function works with diffrent Logic opperators

```python
week.loc[week.Day_name == 'Thursday'].head(1) #(week.Day_name is equivalent to week['Day_name'])
```

# Here we locate the Dayname with Active power above 2.0

```python
week.loc[(week.Day_name == 'Thursday') & (week.Global_active_power > 2.0)].head(1) 
```
